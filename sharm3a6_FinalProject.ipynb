```python
##Importing required libraries
import pandas as pd
import csv
import numpy as np
import matplotlib.pyplot as plt
```

<div class="alert-info">
    Step 1: Import the first dataset- SharkTank Overview (In this dataset, there is information regarding the season number, episode number, company which pitched their idea, idea, Deal proposed or not, deal recieved, rejected, number of presenters in a pitch, attendance of sharks. 
</div>

```python
overview = pd.read_csv("SharkTankOverview.csv")
overview.info()
```

```python
overview.head()
```

<div class="alert-info">
    Step 2: Replace null values with 0 in the Reject offer column
</div>

```python
#Replacing NaN values with 0

overview.fillna(value=0, inplace=True)
overview.head()
```

<div class="alert-info">
    Step 3: Import second dataset : SharkTank Details (In this dataset, each episode is mapped to multiple pitches, the company who present, their idea and each Shark's deal decision
</div>

```python
#Importing the second dataset with Deal decisions

details = pd.read_csv("SharkTankDetails.csv")
details.info()
```

```python
details.head()
```

<div class="alert-info">
    Step 4: Replacing all null values with 0 in the deal decision columns
</div>

```python
#Replacing NaN values with 'N' indicating No deal 

details.fillna(value= 'N', inplace= True)
details.head()
```

<div class="alert-info">
    Step 5: Merge both the datasets with inner join on their Episode and Idea columns, we are performing inner join since we want to analyse the dataset as a whole, with pitches which have been evaluated by the sharks
</div>

```python
#Performing inner join on Episode and Idea columns respectively 

df_full = pd.merge(overview, details, how='inner', left_on = ['Episode', 'Idea'], right_on = ['episode', 'idea'])
df_full.info()
```

```python
df_full.head()
```

<div class="alert-info">
    Step 6: Drop duplicate columns and set index in order to clean the dataset 
</div>

```python
#Dropping the columns named Episode, Company, Idea and Deal since they are duplicate and might confuse the analysis

df_full.drop(['Episode', 'company', 'Idea', 'deal'], axis=1, inplace=True)
```

```python
#Set Season, Episode and Pitch number as the index

df_full = df_full.set_index(['Season','episode','pitch_no'])
df_full.head()
```

<div class="alert-info">
    Step 7: Performing Analysis 
</div>

<div class="alert-warning">
Analysis 1: Determine the Highest Deal bagged, to understand maximum show budget for an episode. 
</div>

```python
#Analysis 1 : Determine Highest deal amount bagged
df_full.nlargest(1, 'Deal_Amount (INR Lakhs)')
```

<div class="alert-success">
    Result obtained: The nlargest query returned one largest entry in the Deal_Amount column with amount of 150Lakhs INR with 5% equity, indicating following information- 
The highest deal was bagged in Episode 17 by Aas Vidyalaya company for their EdTech App Idea. There were 2 presenters comprising 1 female and 1 male presenters, out of 7 Sharks 3 offered the deal with each contributing 50Lakhs.
</div>

<div class="alert-warning">
    Analysis 2: Determine the Easiest and Toughest Shark to impress by visualizing a bar graph indicating each Sharks' number of Yes and No 
</div>

```python
#Collect in a dataframe columns regarding Judges' decision and corresponding Season, Episode and Pitch number 

df_judges = df_full[['ashneer', 'namita','anupam','vineeta','aman','peyush','ghazal']]
df_judges
```

```python
#Count the number of Yes and No for each judge 

df_count = df_judges.apply(pd.value_counts)
df_count
```

```python
#Transpose the dataframe to plot a graph in the next step

df_judges_t = df_count.transpose()
df_judges_t
```

```python
#Plotting a comparative bar graph with N indicating a No decision and Y indicating a Yes decision for all the judges 

df_judges_t.plot(kind='bar', color=['r','g'])
plt.legend(loc='upper left', frameon=False)
plt.ylabel('Count')
plt.xlabel('Judges')
plt.title('Count of total Yes or No decisions')
```

<div class="alert-success">
Results obtained: From the bar graph following information is retrieved- 
1. Ghazal is a Toughest Shark to impress since she has the highest No decisions
2. Aman is an Easiest Shark to impress as his Yes decision count is the highest compared to rest 
</div>

<div class="alert-warning">
    Analysis 3: Determine the relationship between number of presenters and deal success 
</div>

```python
#Plotting a pie chart filtered with pitches which were offered a deal, and visualizing the number of presenters involved in those pitches 

pie_chart = df_full[df_full.Deal==1]['Number of presenters'].value_counts().plot.pie(autopct = '%1.0f%%', 
                                                            title = 'Deal success share with number of presenters')
```

<div class="alert-warning">
    Analysis 4: Determine the relationship between presenter's gender and deal success 
</div>

```python
#Plotting a pie chart filtered with pitches which were offered a deal, and visualizing the number of male presenters and female presenters
#involved in those pitches 

df_full[df_full.Deal==1]['No of male presenters'].value_counts().plot.pie(autopct = '%1.0f%%', 
                        title = 'Deal success share with number of male presenters')

plt.show()


df_full[df_full.Deal==1]['No of female presenters'].value_counts().plot.pie(autopct = '%1.0f%%', 
                        title = 'Deal success share with number of female presenters')

plt.show()
```

<div class="alert-warning">
    Analysis 5: Determine the trend of a pitch being recieved/ rejected over the episodes
</div>

```python
#Created two different dataframes collecting number of pitches offered a deal or being rejected grouped by epsiode number 

df_trend_receive= df_full[df_full['Receive_Offer']== 1].groupby(['episode'])['Receive_Offer'].count()
df_trend_reject= df_full[df_full['Reject_Offer']== 1].groupby(['episode'])['Reject_Offer'].count()


#Merging two dataframes on Episode and replacing null values with 0
df_trend = pd.merge(df_trend_receive, df_trend_reject, on='episode', how='outer')
df_trend.fillna(value= '0', inplace= True)

#Converting the dataframe to integer type object
df_trend = df_trend.astype(int)
df_trend
```

```python
#Plotting a line graph indicating the Episode and number of pitches accepted vs rejected

ax = plt.gca()
df_trend.plot(kind = 'line', y = ['Receive_Offer', 'Reject_Offer'], ax=ax)
plt.xticks(np.arange(1,37,1))
plt.xticks(rotation=90)
plt.yticks(np.arange(1,4,1))
plt.title('Receive/Reject offer count Vs Episode')
plt.xlabel('Episodes')
plt.ylabel('Count')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')

plt.tick_params(axis='x', which='major', labelsize=9)


```

<div class="alert-success">
    Result obtained: A line graph with Y axis as count of pitches in each episode and X axis as Episode 
</div>
